```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Question 1
Symbol="TSLA"
data=yf.Ticker(Symbol)
tesla_data=data.history(period='max')
print(tesla_data.head())
```

```python
#Question 2
financials = data.financials
tesla_revenue = financials.loc['Total Revenue']
print(tesla_revenue.tail())
```

```python
#Question 3
data2=yf.Ticker('GME')
gme_data=data2.history(period='max')
print(gme_data.head())
```

```python
#Question 4
financials2 = data2.financials
gme_revenue = financials2.loc['Total Revenue']
print(gme_revenue.tail())
gme_revenue_df = pd.DataFrame(gme_revenue)
```

```python
#Question 5
def make_graph(df, y_column, title):
    df.plot(y=y_column, kind='bar', title=title)
    plt.ylabel(y_column)
    plt.xlabel('Year')
    plt.show()

tesla_revenue_df = pd.DataFrame(tesla_revenue)
make_graph(tesla_revenue_df, 'Total Revenue', 'Tesla Revenue Over Years')
```

```python
#Question 6
gme_revenue_df = pd.DataFrame(gme_revenue)
make_graph(gme_revenue_df, 'Total Revenue', 'GME Revenue Over Years')
```
